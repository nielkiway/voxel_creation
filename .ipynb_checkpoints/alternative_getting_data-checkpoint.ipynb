```python
import numpy as np 
import h5py
import pandas as pd
```

```python
def get_2D_data_from_h5_filtered_np(h5_path, part_name, Slice_name):
    #opening h5 and getting the data
    h5 = h5py.File(h5_path)                                          
    Y_Axis = np.array(h5[part_name][Slice_name]['Y-Axis'][:]).astype(int)
    Area = np.array(h5[part_name][Slice_name]['Area'][:]).astype(int)
    Intensity = np.array(h5[part_name][Slice_name]['Intensity'][:]).astype(int)
    X_Axis = np.array(h5[part_name][Slice_name]['X-Axis'][:]).astype(int)
    h5.close()
    
    X_Axis_size = X_Axis.size
    Y_Axis_size = Y_Axis.size
    Area_size = Area.size
    Intensity_size = Intensity.size

    #if dimensions aren't equal the following code block is entered
    if not X_Axis_size == Y_Axis_size == Area_size == Intensity_size:

        #determine the lowest value among the different sizes
        size_arr = np.array([X_Axis_size, Y_Axis_size, Area_size, Intensity_size])
        min_size = size_arr.min()

        if X_Axis_size != min_size:
            diff_size_x = X_Axis_size - min_size #calculating the difference between the actual value and the minimum and substracting it from the array
            X_Axis_new = np.delete(X_Axis, -diff_size_x)
            X_Axis = X_Axis_new
            X_Axis_size = X_Axis.size

        if Y_Axis_size != min_size:
            diff_size_y = Y_Axis_size - min_size
            Y_Axis_new = np.delete(Y_Axis, -diff_size_y)
            Y_Axis = Y_Axis_new
            Y_Axis_size = Y_Axis.size

        if Area_size != min_size:
            diff_size_area = Area_size - min_size
            Area_new = np.delete(Area, -diff_size_area)
            Area = Area_new
            Area_size = Area.size

        if Intensity_size != min_size:
            diff_size_intensity = Intensity_size - min_size
            Intensity_new = np.delete(Intensity, -diff_size_intensity)
            Intensity = Intensity_new
            Intensity_size = Intensity.size
                
        #by reducing all the dimensions to the minimum equal dimensions are guaranteed
        #there is a risk of deleting more than just one datapoint without noticing -> maybe add an alert after more than 5(?) while iterations
    print(str(X_Axis_size)+ ' datapoints found')    
    xy_combos = np.stack((X_Axis, Y_Axis), axis=-1)
    xy_combos_unique = np.unique(xy_combos,axis=0)
    print(str(xy_combos_unique.shape[0]) + ' combinations of x and y found')
    
    #filtering out the datapoints where area and intensity are =0
    area_zeros = np.where(Area== 0) 
    result_intensity = np.where(Intensity==0)
    Zero_area_intensity_indices = np.intersect1d(area_zeros, result_intensity) #array of indices where area AND intensity are = 0
    
    #deleting all the datapoints where area AND intensity are = 0
    X_Axis_wo_Zeros = np.delete(X_Axis, Zero_area_intensity_indices)
    Y_Axis_wo_Zeros = np.delete(Y_Axis, Zero_area_intensity_indices)
    Area_wo_Zeros = np.delete(Area, Zero_area_intensity_indices)
    Intensity_wo_Zeros = np.delete(Intensity, Zero_area_intensity_indices)
    
    xy_combos_wo_only_zeros = np.stack((X_Axis_wo_Zeros, Y_Axis_wo_Zeros), axis = -1)
    xy_combos_wo_only_zeros_unique, unique_indices = np.unique(xy_combos_wo_only_zeros,axis=0, return_index = True)
    #unique_indexes is an array of all the indices with corresponding x-y-combinations that occur only once
    print(str(xy_combos_wo_only_zeros_unique.shape[0]) + ' unique combinations of x and y found (area ==0 and intensity == 0 removed)')
    #get the corresponding area and intensity values
    area_wo_only_zeros_unique = Area_wo_Zeros[unique_indices]
    intensity_wo_only_zeros_unique = Intensity_wo_Zeros[unique_indices]
    X_Axis_wo_only_zeros = X_Axis_wo_Zeros[unique_indices]
    Y_Axis_wo_only_zeros = Y_Axis_wo_Zeros[unique_indices]
    xy_combo_final = np.stack((X_Axis_wo_only_zeros, Y_Axis_wo_only_zeros), axis=-1)
    
    
    Index_range = np.arange(xy_combos_wo_only_zeros.shape[0])
    indices_of_interest = np.setdiff1d(Index_range, unique_indices)
    print(str(indices_of_interest.shape[0]) + ' positions occur multiple times (area ==0 and intensity == 0 removed)')
          
    #initialize empty arrays for storing the xy_combos and corresponding max. area and max. intensity values
    xy_combo_array = np.empty([0,2],dtype= int)
    area_max_array = np.empty([0,1],dtype= int)
    intensity_max_array = np.empty([0,1],dtype= int)
          
    for index in indices_of_interest:
        print(index)
        #check ob xy combo in array
        if xy_combos_wo_only_zeros[index] not in xy_combo_array:
        #getting the x and y value combination of that particular index
            xy_combo = xy_combos_wo_only_zeros[index]
            #alle indizes von genau dieser Kombi suchen
            indices_of_combo = np.where(xy_combos_wo_only_zeros == xy_combo)[0]
            #über indizes die werte finden + mean oder max berechnen
            area_of_combo = Area_wo_Zeros[indices_of_combo]
            intensity_of_combo = Intensity_wo_Zeros[indices_of_combo]
            max_area_of_combo = np.amax(area_of_combo)
            max_intensity_of_combo = np.amax(intensity_of_combo)
            #xy combo in array und area und intensität auch 
            xy_combo_array = np.vstack((xy_combo_array,xy_combo))
            area_max_array = np.vstack((area_max_array, max_area_of_combo))
            intensity_max_array = np.vstack((intensity_max_array, max_intensity_of_combo))
            # die zu einem großen array machen
    
    #searching for the xy_combo in xy_combos_wo_Zeros_only and change the area 
    for index in np.arange(xy_combo_array.shape[0]):
        print('index = ' + str(index))
        xy_position = xy_combo_array[index]
        index_in_main_array = np.where(xy_combo_final == xy_position)[0]#
        print('index in main array is ' +str(index_in_main_array))
        area_wo_only_zeros_unique[index_in_main_array] = area_max_array[index]
        intensity_wo_only_zeros_unique[index_in_main_array] = intensity_max_array[index]
    
    X_Axis_final = np.hsplit(xy_combo_final, 2)[0]    
    Y_Axis_final = np.hsplit(xy_combo_final, 2)[1] 
    Area_final = area_wo_only_zeros_unique
    Intensity_final = intensity_wo_only_zeros_unique  
          
    help_arr = np.column_stack((X_Axis_final, Y_Axis_final, Area_final, Intensity_final))
    df_raw = pd.DataFrame(help_arr, columns=['x','y','area','intensity'])
    print(xy_combo_array) 
    print(area_max_array)
    print(intensity_max_array)
    return(df_raw)
    
```
