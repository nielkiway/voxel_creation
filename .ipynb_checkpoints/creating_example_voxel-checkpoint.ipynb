```python
import h5py
import numpy as np
import math
import pandas as pd
```

```python
voxel_size = 5
#z-dimension of the voxel for first tries not relevant
```

```python
#Berechnung der Anzahl an Voxeln (hier simpler Fall)
def get_number_voxel (max_x_part, max_y_part, voxel_size):
    n_voxels_x = math.ceil(int(max_x_part/voxel_size))
    n_voxels_y = math.ceil(int(max_y_part/voxel_size))
    return n_voxels_x, n_voxels_y #tuple is returned 

#n_voxels_x = int(20/voxel_size)
#n_voxels_y = int(20/voxel_size) #muss noch geändert und flexibilisiert werden. 20 wegen x-Länge und y-Länge
#print(n_voxels_x)
#print(n_voxels_y)
```

```python
num_voxels = get_number_voxel(20,20,voxel_size)
```

```python
#getting the 2D-data from the hdf5 file and saving it in a pd DataFrame

def get_2D_data_from_h5(h5_path, part_name, Slice_name):
    with h5py.File(h5_path,'r') as h5:
        X_Axis = h5[part_name][Slice_name]['X_Axis']
        Y_Axis = h5[part_name][Slice_name]['Y_Axis']
        Area = h5[part_name][Slice_name]['Area']
        Intensity = h5[part_name][Slice_name]['Intensity']
        
        help_arr = np.column_stack((X_Axis, Y_Axis, Area, Intensity))
        df = pd.DataFrame(help_arr, columns=['x','y','Area','Intensity'])
        return df
```

```python
df = get_2D_data_from_h5("/home/jan/Documents/Voxel_Erstellung/example_hdf_3.hdf",'part_1','Slice_1')
```

```python
#-----------------------------------------------------------------------------------------------
# automatisiertes Befüllen mehrerer Voxel
# direkt hdf-Struktur mit Voxeln anlegen mit Voxelnamen als 'Group' --> Benennungsproblem umgangen
# so voxel direkt identifizierbar und einer bestimmten Stelle im Bauteil zuordenbar
```

```python
#Voxel 2D erzeugt und befüllt für nur eine Schicht in allen Voxels dieser Schicht

for n_vox_y_init in range(num_voxels[0]): #iterating over number of voxels in y-direction
    for n_vox_x_init in range(num_voxels[1]):#iterating over number of voxels in x-direction 
        
        array = np.zeros([voxel_size,voxel_size]) #creating an empty array of dimensions voxel_size*voxel_size
        for i in range(voxel_size*n_vox_x_init, voxel_size*(n_vox_x_init+1)): #iterating over x 
            for j in range(voxel_size*n_vox_y_init,voxel_size*(n_vox_y_init+1)): #iterating over y
                
                #finding the area value for a certain point in the part-data-dataframe and allocating it to a position in the array
                area_i = df.loc[(df['x'] == i) & (df['y'] == j)]
                array[i-n_vox_x_init*voxel_size][j-n_vox_y_init*voxel_size] = area_i['area']
                
        with h5py.File("voxels.hdf5", "a") as voxel_hdf:
            #creating a voxel with the numbers of voxels in both direction in its name and filling it with data
            voxel_hdf.create_group('voxel_{}_{}'.format(n_vox_x_init,n_vox_y_init))
            voxel_hdf['voxel_{}_{}'.format(n_vox_x_init,n_vox_y_init)].create_dataset('X-Axis',data = np.repeat(np.arange(0,voxel_size,1),voxel_size))
            voxel_hdf['voxel_{}_{}'.format(n_vox_x_init,n_vox_y_init)].create_dataset('Y-Axis',data = np.tile(np.arange(0,voxel_size,1),voxel_size))
            voxel_hdf['voxel_{}_{}'.format(n_vox_x_init,n_vox_y_init)].create_dataset('Area', data = array.flatten())
        #print(array)       
# y-Achse entspricht horizontal, x-Achse entspricht vertikal
```

```python
#modularizing this part of the loop (transfering it to a function)
#input_variables: voxel_size, num_voxels_x, num_voxels_y
#output: array filled with area values

def fill_2D_voxel (voxel_size, num_voxels_x, num_voxels_y, dataframe):
    array = np.zeros([voxel_size,voxel_size]) #creating an empty array of dimensions voxel_size*voxel_size
    for i in range(voxel_size*num_voxels_x, voxel_size*(num_voxels_x+1)): #iterating over x 
        for j in range(voxel_size*num_voxels_y,voxel_size*(num_voxels_y+1)): #iterating over y

            #finding the area value for a certain point in the part-data-dataframe and allocating it to a position in the array
            area_i = dataframe.loc[(dataframe['x'] == i) & (dataframe['y'] == j)]
            array[i-num_voxels_x*voxel_size][j-num_voxels_y*voxel_size] = area_i['area']
            
    return array 
```

```python
def fill_2D_voxel_area (voxel_size, num_voxels_x, num_voxels_y, dataframe):
    array_area = np.zeros([voxel_size,voxel_size]) #creating an empty array of dimensions voxel_size*voxel_size
    for i in range(voxel_size*num_voxels_x, voxel_size*(num_voxels_x+1)): #iterating over x
        for j in range(voxel_size*num_voxels_y,voxel_size*(num_voxels_y+1)): #iterating over y

            #finding the area value for a certain point in the part-data-dataframe and allocating it to a position in the array
            area_i = dataframe.loc[(dataframe['x'] == i) & (dataframe['y'] == j)]
            array_area[i-num_voxels_x*voxel_size][j-num_voxels_y*voxel_size] = area_i['Area']

    return array_area
```

```python
def fill_2D_voxel_intensity (voxel_size, num_voxels_x, num_voxels_y, dataframe):
    array_intensity = np.zeros([voxel_size,voxel_size]) #creating an empty array of dimensions voxel_size*voxel_size
    for i in range(voxel_size*num_voxels_x, voxel_size*(num_voxels_x+1)): #iterating over x
        for j in range(voxel_size*num_voxels_y,voxel_size*(num_voxels_y+1)): #iterating over y

            #finding the area value for a certain point in the part-data-dataframe and allocating it to a position in the array
            area_i = dataframe.loc[(dataframe['x'] == i) & (dataframe['y'] == j)]
            array_intensity[i-num_voxels_x*voxel_size][j-num_voxels_y*voxel_size] = area_i['Intensity']

    return array_intensity
```
