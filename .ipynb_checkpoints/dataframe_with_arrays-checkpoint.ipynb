```python
for index in indices_of_interest:
    print(index)
    #check ob xy combo in array
    if xy_combos_wo_zeros_only[index] not in xy_combo_array:
    #getting the x and y value combination of that particular index
        xy_combo = xy_combos_wo_zeros_only[index]
        #alle indizes von genau dieser Kombi suchen
        indices_of_combo = np.where(xy_combos_wo_zeros_only == xy_combo)[0]
        #über indizes die werte finden + mean oder max berechnen
        area_of_combo = Area_wo_Zeros[indices_of_combo]
        intensity_of_combo = Intensity_wo_Zeros[indices_of_combo]
        max_area_of_combo = np.amax(area_of_combo)
        max_intensity_of_combo = np.amax(intensity_of_combo)
        #xy combo in array und area und intensität auch 
        xy_combo_array = np.vstack((xy_combo_array,xy_combo))
        area_max_array = np.vstack((area_max_array, max_area_of_combo))
        intensity_max_array = np.vstack((intensity_max_array, max_intensity_of_combo))
        # die zu einem großen array machen
```
