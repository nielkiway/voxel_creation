```python
import h5py
import numpy as np
import pandas as pd
import math
import re
```

```python
#here it is checked which slices contain information 
```

```python
list_z_param = []
with h5py.File('/home/jan/Documents/Klein_Datentransfer/HDF/BJ_Superlativ_QualiPro.h5','a') as h5:
    key_list = h5['0_00097_ZP1_Z_cls'].keys()
    
    for key in key_list:
       #shape suchen 
        shape = h5['0_00097_ZP1_Z_cls'][key]['Area'].shape[0]
        if shape > 0: 
            list_z_param.append(key)
```

```python
list_p_param = []
with h5py.File('/home/jan/Documents/Klein_Datentransfer/HDF/BJ_Superlativ_QualiPro.h5','a') as h5:
    key_list = h5['0_00099_ZP1_P_3_cls'].keys()
    
    for key in key_list:
       #shape suchen 
        shape = h5['0_00099_ZP1_P_3_cls'][key]['Area'].shape[0]
        if shape > 0: 
            list_p_param.append(key)
```

```python
Training_hdf = h5py.File('/home/jan/Documents/Trainingsdaten/HDF5_full_part.h5', "w")
Training_hdf.close()
```

```python
with h5py.File('/home/jan/Documents/Trainingsdaten/HDF5_full_part.h5','a') as h5:
    h5.create_group("ZP1_combined")
    for slice_num in list_z_param:
        #the following block transforms the uneven layer numbers in normal numbers
        slice_num_int = int(re.search(r'\d+', slice_num).group())
        slice_num_normal = math.trunc(slice_num_int/2)
        slice_name_normal = 'Slice' + str("{:05d}".format(slice_num_normal))  
        
        h5["ZP1_combined"].create_group(slice_name_normal)
        with h5py.File('/home/jan/Documents/Klein_Datentransfer/HDF/BJ_Superlativ_QualiPro.h5','a') as h5_2:
            h5["ZP1_combined"][slice_name_normal].create_dataset('Area',data = h5_2['0_00097_ZP1_Z_cls'][slice_num]['Area'])
            h5["ZP1_combined"][slice_name_normal].create_dataset('Intensity',data = h5_2['0_00097_ZP1_Z_cls'][slice_num]['Intensity'])
            h5["ZP1_combined"][slice_name_normal].create_dataset('X-Axis',data = h5_2['0_00097_ZP1_Z_cls'][slice_num]['X-Axis'])
            h5["ZP1_combined"][slice_name_normal].create_dataset('Y-Axis',data = h5_2['0_00097_ZP1_Z_cls'][slice_num]['Y-Axis'])
            
    for slice_num in list_p_param:
        #the following block transforms the uneven layer numbers in normal numbers
        slice_num_int_p = int(re.search(r'\d+', slice_num).group())
        slice_num_normal_p = math.trunc(slice_num_int_p/2)
        slice_name_normal_p = 'Slice' + str("{:05d}".format(slice_num_normal_p))
        
        h5["ZP1_combined"].create_group(slice_name_normal_p)
        with h5py.File('/home/jan/Documents/Klein_Datentransfer/HDF/BJ_Superlativ_QualiPro.h5','a') as h5_2:
            h5["ZP1_combined"][slice_name_normal_p].create_dataset('Area',data = h5_2['0_00099_ZP1_P_3_cls'][slice_num]['Area'])
            h5["ZP1_combined"][slice_name_normal_p].create_dataset('Intensity',data = h5_2['0_00099_ZP1_P_3_cls'][slice_num]['Intensity'])
            h5["ZP1_combined"][slice_name_normal_p].create_dataset('X-Axis',data = h5_2['0_00099_ZP1_P_3_cls'][slice_num]['X-Axis'])
            h5["ZP1_combined"][slice_name_normal_p].create_dataset('Y-Axis',data = h5_2['0_00099_ZP1_P_3_cls'][slice_num]['Y-Axis'])        
```
